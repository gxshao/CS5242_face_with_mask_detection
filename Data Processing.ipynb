# Data Processing

```python
import os
import shutil
from PIL import Image
import random
import torch

root_dir = os.getcwd() + '/data_processing/'
face_dir = root_dir + 'label_face/'
non_face_dir = root_dir + 'label_non_face/'
data_dir = root_dir + 'data/'

print("Root dir: ", root_dir, "\nFace dir: ", face_dir, " \nnon-face dir:", non_face_dir)
print("Data dir: ", data_dir)
```

### 1. Resize

```python
def resize(target_dir):
    count = 0
    for root, dirs, files in os.walk(target_dir):
        for filename in files:
            print("processing: ", filename)
            try:
                image = Image.open(target_dir + filename)
                n = image.resize((100, 100))
                n.save(target_dir + filename )
            except:
                print("Resize failed, ready to remove")
                os.remove(filename)
            count += 1
    print("Resize finished: ", count)
resize(face_dir)
resize(non_face_dir)
```

### 2. Augmentation

```python
def augmentation(target_dir):
    count = 0
    for root, dirs, files in os.walk(target_dir):
        for filename in files:
            print("processing: ", filename)
            try:
                image = Image.open(target_dir + filename)
                deg = random.randint(10, 15)
                n = image.rotate(deg)
                n.save(target_dir + "aug_" + str(deg) + "_" + filename)
                deg = random.randint(345, 350)
                n = image.rotate(deg)
                n.save(target_dir + "aug_" + str(deg) + "_" + filename)
                count += 1
            except:
                print("Skipping: ", filename)
    print("Resize finished: ", count, " Target dir: ", target_dir)
    
augmentation(non_face_dir)
augmentation(face_dir)
```

### 3. Shuffle and label the data

```python
def generate_labels():
    # Calculate the files sum
    count_face_files = len(os.listdir(face_dir))
    count_non_face_files = len(os.listdir(non_face_dir))
    total_files = count_face_files + count_non_face_files
    print("Face with mask files: ", count_face_files, " non-face files:", count_non_face_files, " in total: ", total_files)

    label_face = torch.ones(count_face_files)
    label_non_face = torch.zeros(count_non_face_files)
    labels = torch.cat((label_non_face, label_face), 0).long()
    torch.save(labels, root_dir + 'labels.pt')
    print("Label writting success")
    return count_face_files, count_non_face_files

# Rename the data
def rename(target_dir, name_offset):
    for root, dirs, files in os.walk(target_dir):
        name = name_offset
        for filename in files:
            shutil.copyfile(target_dir + filename, data_dir + str(name) + '.jpg')
            name += 1
        print("Renaming done")

c_face, c_non_face = generate_labels()
rename(non_face_dir, 0)
rename(face_dir, c_non_face)
```

### 4. Validate data set and labels

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

total_data_files = len(os.listdir(data_dir))
print("Find data: ", total_data_files)

labels = torch.load(root_dir + 'labels.pt')
rand = random.randint(0, total_data_files)
print("index:", rand, " label:", labels[rand].item())
img = mpimg.imread(data_dir + str(rand) + '.jpg')
imgplot = plt.imshow(img)
plt.show()
```
