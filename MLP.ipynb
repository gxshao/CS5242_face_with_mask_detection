# MLP

```python
import sys, os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
import time
import torch.utils.data as data
from PIL import Image
import torchvision.transforms as transforms
import cv2

root_dir = os.getcwd() + '/finaldata'

train_dir = root_dir + '/train/'
# val_dir = root_dir + '/val/'
test_dir = root_dir + '/test/'

print(root_dir)
print(train_dir)
# print(val_dir)
print(test_dir)
```

```python
class FaceSet(data.Dataset):
    def __init__(self,root,train,test,transform,split):
        self.train=train
        self.test=test
        self.transform=transform
        imgs=[os.path.join(root,f) for f in os.listdir(root)]
        if self.test:
            imgs=sorted(imgs,key=lambda x:int(x.split('/')[-1].split('.')[0].split('_')[0]))
            self.imgs=imgs
        else:
            imgsnum=len(imgs)
            imgs=sorted(imgs,key=lambda x:int(x.split('/')[-1].split('.')[0].split('_')[0]))
            if self.train:
                self.imgs=imgs[:int(split*imgsnum)]
            else:
                self.imgs=imgs[int(split*imgsnum):]
        
    def __getitem__(self,index):
        imgpath=self.imgs[index]
        im=Image.open(imgpath)
        if im.mode == "RGBA":
            r, g, b, a=im.split ()
            im=Image.merge ("RGB", (r, g, b))
        elif im.mode!="RGB":
            im=im.convert ("RGB")
        im=self.transform(im)
        symbol = imgpath.split('/')[-1].split('.')[0].split('_')[-1]
        if symbol == 'face': label = 2
        elif symbol == 'nomask': label = 1
        elif symbol == 'nonface': label = 0
        return im,label
    def __len__(self):
        return len(self.imgs)
```

```python
imgsize=100
transform_train=transforms.Compose([ 
    transforms.Resize((100,100)),
    transforms.RandomCrop((imgsize,imgsize)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
])
transform_test=transforms.Compose([ 
    transforms.Resize((imgsize,imgsize)),
    transforms.ToTensor(),
])

for fn in os.listdir(train_dir):
    im=Image.open(train_dir + fn)
    print(transform_train(im).size())
    
trainset=FaceSet(root=train_dir,train=True,test=False,transform=transform_train,split=0.85)
# valset=FaceSet(root=val_dir,train=False,test=False,transform=transform_test,split=0.85)
valset=FaceSet(root=train_dir,train=False,test=False,transform=transform_test,split=0.85)
testset=FaceSet(root=test_dir,train=False,test=True,transform=transform_test,split=1)
```

```python
device = torch.device("cpu")
```

```python
class mlp(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(mlp , self).__init__()
        self.layer1 = nn.Linear( input_size, hidden_size1 , bias=False)
        self.layer2 = nn.Linear( hidden_size1, hidden_size2 , bias=False)
        self.layer3 = nn.Linear( hidden_size2, output_size , bias=False)
        
    def forward(self, x):
        x = self.layer1(x)
        x = F.relu(x)
        x = self.layer2(x)
        x = F.relu(x)
        x = self.layer3(x)
        return x

def get_error(scores , labels):
    bs=scores.size(0)
    predicted_labels = scores.argmax(dim=1)
    indicator = (predicted_labels == labels)
    num_matches=indicator.sum()
    
    return 1-num_matches.float()/bs  

def eval_on_test_set(loader):
    running_error=0
    num_batches=0
    for batch_idx,(img,label) in enumerate(loader):
        img=img.view(img.size()[0],-1)
        img=img.to(device)
        scores=net(img) 
        label=label.to(device) 
        error = get_error( scores , label)
        running_error += error.item()
        num_batches+=1
    total_error = running_error/num_batches
    print( 'test error  = ', total_error*100 ,'percent')
```

```python
net = mlp(imgsize**2*3, imgsize**2, imgsize, 3)
net = net.to(device)
criterion = nn.CrossEntropyLoss()
bs = 10
trainloader=torch.utils.data.DataLoader(trainset,batch_size=bs,shuffle=True,num_workers=0)
valloader=torch.utils.data.DataLoader(valset,batch_size=bs,shuffle=False,num_workers=0)
testloader=torch.utils.data.DataLoader(testset,batch_size=bs,shuffle=False,num_workers=0)
```

```python
import torch
print(torch.__version__)
print(torch.cuda.is_available())
```

```python
start = time.time()
lr = 0.1
for epoch in range(100):
    if epoch % 10 == 0 and epoch > 10: 
        lr = lr / 1.5
    optimizer=torch.optim.SGD(net.parameters(), lr = lr)
    running_loss=0
    running_error=0
    num_batches=0
    for batch_idx,(img,label) in enumerate(trainloader):
        optimizer.zero_grad()
        inputs = img.view(img.size()[0],-1)
        inputs.requires_grad_()
        inputs=inputs.to(device)
        scores=net(inputs) 
        label=label.to(device)
        loss =  criterion( scores , label) 
        loss.backward()
        optimizer.step()
        torch.cuda.empty_cache()
        running_loss += loss.detach().item()
        error = get_error( scores.detach() , label)
        running_error += error.item()
        num_batches+=1

    total_loss = running_loss/num_batches
    total_error = running_error/num_batches
    elapsed_time = time.time() - start
    if epoch % 10 == 0 : 
        print(' ')
        print('epoch=',epoch, ' time=', elapsed_time,
              ' loss=', total_loss , ' error=', total_error*100 ,'percent lr=', lr)
        eval_on_test_set(valloader)
```
